```python
def singularValue(data,boundary):
    data = np.array(data)
    mean = int(data.mean())
    for i in range(len(data)):
        if data[i] > boundary:
            data[i] = mean
    return data
```

```python
data_train['卧室数量'] = singularValue(np.array(data_train['卧室数量']),5)
data_train['厅的数量'] = singularValue(np.array(data_train['厅的数量']),4)
data_train['卫的数量'] = singularValue(np.array(data_train['卫的数量']),3)
data_train.describe()
```

```python
all_data_missing = (data_train.isnull().sum() / len(data_train)) * 100
all_data_missing = all_data_missing.drop(all_data_missing[all_data_missing == 0].index).sort_values(ascending=False)
missing_data = pd.DataFrame({'Missing Percentage':all_data_missing})
missing_data
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(10, 6))
plt.xticks(rotation='90')
sns.barplot(x=missing_data.index, y=missing_data['Missing Percentage'])
plt.xlabel('Features',fontproperties = font)
plt.ylabel('Percent of missing values',fontproperties = font)
plt.title('Percent missing data by feature', fontproperties = font)
plt.show()
```

```python
for col in ['装修情况','居住状态','出租方式']:
    data_train[col] = data_train[col].fillna('None')
```

```python
for col in ['距离','地铁站点','地铁线路']:
    data_train[col] = data_train[col].fillna(0)
```

```python
for col in ['小区房屋出租数量','位置','区']:
    data_train[col] = data_train[col].fillna(data_train[col].median())
```
